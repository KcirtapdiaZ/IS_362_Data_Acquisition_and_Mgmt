Start with importing the pandas library which will allow us to read the csv file. We use the import to add the pandas module

```python
import pandas as pd
```

Need to load the csv into the pandas dataframe. Pulling from my GitHub repository. Define the variable as df and use the read_csv() function to load the csv into pandas. I added the extra df line to display the data

```python
df = pd.read_csv("https://github.com/KcirtapdiaZ/IS_362_Data_Acquisition_and_Mgmt/raw/main/Week4-Project1.csv")
df
```

Filter the dataset to show only delayed flights to make it easier to calculate the delays for each airline. This creates a new DataFrame that only contains the delayed flights. To simplify later code I filter now

```python
delays = df[df["Status"] == "Delayed"]
delays
```

In order to calculate the total delays by airline, need to group the delayed flights by each airline. This is make it easier to compare which airline has the most delays overall. Best way to add up the number of delays is to group. By using groupby we can combine the data by category

```python
delays_by_airline = delays.groupby("Airline")["Count"].sum()
delays_by_airline
```

To calculate the overall delay percentage we need to first calculate the total number of flights for each airline. Both On Time and Delayed. I used groupby again for code consistancy

```python
total_flights = df.groupby("Airline")["Count"].sum()
total_flights
```

We can now calculate the delay percentage of each airline. We divide variable "delays_by_airline" by "total_flights" and multiply by 100.

```python
delay_percentages = (delays_by_airline / total_flights) * 100
delay_percentages
```

Analyzing the data of the delays between each airline, the percentages are a lot closer than you may think just by baseing your conclusion off of which airline had more total flights or more delays. The percentage shows you a better picture, becasue although AM West had more delays at 787 compared to 516, they also had substancially more total flights at 7225 compared to 3981. Because the delay percentage is so close maybe factors like destination, flight length, or type of plane used should be factored in to determine a true airline reliability. If AM West didnt have 2300 more flights the delay percentage may be higher which would reverse the conclusion of AM West is more reliable because they have less delay percentage.
