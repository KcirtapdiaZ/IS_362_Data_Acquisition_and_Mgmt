# The Zip () in Python is used to combine multiple iterables in to a single iterable of tuples. When dealing with iterables of varying lenghts this can function can be tricky and trying to unzip values can be a bit difficult.

```python
# Two List

names = ["Harry", "Ron", "Hermione"]
ages = [17, 17, 18]

# Use Zip function to combine lists
combined = list(zip(names, ages))
print(combined)
```

# Dealing with iterables of varying lenghts. The Zip function will stop pairing at the shortest iteration.

```python
names = ["Harry", "Ron", "Hermione", "Draco"]
ages = [17, 17, 18]

# Use Zip function to combine lists
combined = list(zip(names, ages))
print(combined)
```

# Unzip values

```python
# Zipped object

zipped = list(zip(names, ages))
print("Zipped:", zipped)

#Unzipping object
names_unzip, ages_unzip = zip(*zipped)
print("Names:", names_unzip)
print("Ages:", ages_unzip)
```

# The use cases for this are to combine multiple list for iteration, creating dictionaries from list, parallel interation using loops

```python
# Use to create dictionary

names = ["Harry", "Ron", "Hermione"]
ages = [17, 17, 18]

name_age_dict = dict(zip(names, ages))
print(name_age_dict)
```
