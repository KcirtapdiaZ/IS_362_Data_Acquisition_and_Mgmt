Using the third dataset List of NBA players Stats

The objective will be to turn this wide format into long format to see player stats more easily 

Dataset Source: https://www.nba.com/stats/players/traditional?sort=PTS&dir=-1%29

First we will import pandas

```python
import pandas as pd
```

Next we will load the csv. I uploaded the csv to my GitHub so I can pull from there instead of the local PC. I will define the variable as df and use the read_csv() function to load the csv into pandas. I will us df line to show the data. 

```python
df = pd.read_csv('https://github.com/KcirtapdiaZ/IS_362_Data_Acquisition_and_Mgmt/raw/main/IS362_Project2_Dataset3.csv')
df
```

Since we are going to be turing in this in a long format we do want to keep some columns the way they . We will identify those columns now

```python
keep_columns = ["Player", "Team"]
```

Now we'll define the stat columns

```python
stats_to_show = ["PTS", "REB", "AST", "STL", "BLK"]
```

Lastly we are going to need to turn the original columns into rows. we can use the .melt to perform this

```python
df_new = df.melt(id_vars=keep_columns, value_vars=stats_to_show, var_name="Stat", value_name="Value")
df_new
```

Lastly, I think it would be best to group by the player name so that the stats per player are together to easily analyze. I think the .sort_values would be a good function to use

```python
df_new = df_new.sort_values(["Player"])
df_new
```

With this we can easily identify all the main stats for a player, which improves the readability and organization for analysis
