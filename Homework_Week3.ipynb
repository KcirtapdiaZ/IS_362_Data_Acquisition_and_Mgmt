```python
import pandas as pd

# Miles recorded at the end of each day
cumulative_miles = pd.Series(
    [55, 120, 185, 240, 310, 375],
    index = ["Day 1", "Day 2", "Day 3", "Day 4", "Day 5", "Day 6"]
)

# Calculate miles each day
daily_miles = cumulative_miles.diff()
daily_miles[0] = cumulative_miles[0]  # This will set the first day manualy

# Results
cumulative_miles, daily_miles
```
