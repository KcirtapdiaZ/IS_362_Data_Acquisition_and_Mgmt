```python
import pandas as pd

# Create a pandas series of cumulative miles
cumulative_miles = pd.Series(
    [55, 120, 185, 240, 310, 375],
    index = ["Day 1", "Day 2", "Day 3", "Day 4", "Day 5", "Day 6"],
    name = "Cumulative Miles"
)

# Calculate miles each day
daily_miles = cumulative_miles.diff().fillna(cumulative_miles.iloc[0]).astype(int)
daily_miles.name = 'Daily Miles'

#Result
cumulative_miles, daily_miles
```

```python
mileage_df = pd.DataFrame({
    "Cumulative Miles": cumulative_miles,
    "Daily Miles": daily_miles
})

mileage_df
```
