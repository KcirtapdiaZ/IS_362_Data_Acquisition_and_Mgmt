Using the second dataset List of World Series Champions

The objective will be to analyze franchise performance to see how many world series teams have won

Dataset Source: https://en.wikipedia.org/wiki/List_of_World_Series_champions

First we will import pandas library

```python
import pandas as pd
```

Now we'll load the csv. I uploaded the csv to my GitHub so I can pull from there instead of the local PC. I will define the variable as df and use the read_csv() function to load the csv into pandas. I will us df line to show the data.

```python
df = pd.read_csv('https://github.com/KcirtapdiaZ/IS_362_Data_Acquisition_and_Mgmt/raw/main/IS362_Project2_Dataset2.csv')
df
```

In order to determine how many world series a franchise has won, we should clean up the names and replace any old team names with the new ones. This will allow for a easier comparison later

We can see in the Winning team column there are numbers after the names which would make it difficult to group the teams properly so, first we will remove the record values after the team names in the Winning team column

```python
df["Winning team"] = df["Winning team"].str.replace(r"\(.*\)", "", regex=True)
```

We will need to update the names for the following teams:
Dodgers, Giants, Braves, Orioles, Twins, Rangers, Nationals, Astros, Angels, Marlins, Rays, Guardians

```python
team_name_update = {
    "Brooklyn Dodgers": "Los Angeles Dodgers",
    "New York Giants": "San Francisco Giants",
    "Boston Braves": "Atlanta Braves",
    "Milwaukee Braves": "Atlanta Braves",
    "Philadelphia Athletics": "Oakland Athletics",
    "Kansas City Athletics": "Oakland Athletics",
    "St. Louis Browns": "Baltimore Orioles",
    "Washington Senators (1901-1960)": "Minnesota Twins",
    "Washington Senators (1961-1971)": "Texas Rangers",
    "Montreal Expos": "Washington Nationals",
    "Houston Colt .45s": "Houston Astros",
    "California Angels": "Los Angeles Angels",
    "Anaheim Angels": "Los Angeles Angels",
    "Florida Marlins": "Miami Marlins",
    "Tampa Bay Devil Rays": "Tampa Bay Rays",
    "Cleverland Indians": "Cleveland Guardians",
}
```

Now we'll replace the names in the dataset. Since we are only going to calculating how times a frachise has won, we only care about the winning team column

```python
df["Winning team"] = df["Winning team"].replace(team_name_update)
```

Now we'll calculate how many times a team has one by using the .value_counts formula

```python
number_of_championships = df["Winning team"].value_counts()
number_of_championships.head(10)
```
