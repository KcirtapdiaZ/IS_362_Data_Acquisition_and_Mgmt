Using the first dataset Best Selling Video Games

The objective will be to analyze the total sales by publisher to identify which companies have the most best selling video games

Dataset Source: https://en.wikipedia.org/wiki/List_of_best-selling_video_game

First we need to import the pandas library

```python
import pandas as pd
```

Let now load the csv. I uploaded the csv to my GitHub so I can pull from there instead of the local PC. I will define the variable as df and use the read_csv() function to load the csv into pandas. I will us df line to show the data. In order to properly read this, i needed to clear all the non utf8 character in the csv.

```python
df = pd.read_csv('https://github.com/KcirtapdiaZ/IS_362_Data_Acquisition_and_Mgmt/raw/main/IS362_Project2_Dataset1.csv')
df
```

We can see in the publisher column there are similar companies with the same name. I will group these up for better analysis later. We will replace the names so they can be group properly.

```python
publisher_merge = {
    "Nintendo / The Pokemon Company": "Nintendo",
    "Sony Computer Entertainment": "Sony Interactive Entertainment",
    "Re-Logic / 505 Games": "505 Games"
}

df["Publisher(s)"] = df["Publisher(s)"].replace(publisher_merge)
```

To find how many times a publisher appears in the chart I will use a .values_counts formula and create a new variable to hold the data. 

```python
publisher_appearances = df["Publisher(s)"].value_counts()
publisher_appearances.head(50)
```

We can see above that Nintendo appears the most in the data. They have 20 of the top selling video games, which is a lot compared to the runner up at only 5. 

I think it would also be beneficial to calculate which companies sold the most based on publisher as this can also be helpful in analyzing. 

```python
df.groupby("Publisher(s)")["Sales (millions)"].sum().sort_values(ascending=False)
```

We can see that Nintendo is also on the top of this list as well. This would make sense as Nintendo has 20 games in the top selling. We can also deduct that even though Activision has the second most top published games they have substancially less sales.
