For question 3 we need to pull in a different csv file. I will pull it from the same repository as before to make sure the code can be run without needing to save the file locally

```python
weather = pd.read_csv("https://raw.githubusercontent.com/tidyverse/nycflights13/main/data-raw/weather.csv")
weather.head()
```

Question 3 is On February 12th, 2013, which New York area airport had the windiest weather?
In order to find this we will need to do some filtering.

```python
feb122013 = weather[(weather["month"] == 2) & (weather["day"] == 12) & (weather["year"] == 2013)]
feb122013.head()
```

We are looking for the airport with the windiest weather so we will need to sort by the wind speed column.

```python
feb122013.sort_values('wind_speed', ascending=False,)
feb122013.head()
```

Now we can see a value with a wind speed over 1000 which is definately an error. This needs to be removed. I am going to use the < function to ensure I dont pull wind speeds over 100.

I will create a new variable to hold the new data after using the < function

```python
feb122013_clean = feb122013[feb122013['wind_speed'] < 200]
feb122013_clean.head()
```

I could've added the sorting with the new variable to the above code but I figured I'd seperate it so easier visibility. I needed to add the .head method to the end of this line of code otherwise the data would not sort properly again.

```python
feb122013_clean.sort_values(by='wind_speed', ascending=False).head()
```

Looking at the sorted data chart we can see that LGA had the windiest weather on February 12th 2013
