The goal of this project is to display the following in a table: Customer first and last name, track name, and album title from a sqlite database using a pandas dataframe

First we will import the needed libraries -- sqlite3 will allow us to read the sqlite database

```python
import sqlite3
import pandas as pd
```

After importing the libraries, we need to connect to proper chinook database. This database is also loaded into my GitHub reposotory. We'll use the conn function to tell the code which database to connect to

```python
conn = sqlite3.connect("chinook_sqlite.sqlite")
```

The Album table holds the album titles--
The Customer table holds the names of the customers--
The Invoice table links the customers to the purchase--
The InvoiceLine table contains the items that were purchased--
The Track table holds the names of the tracks--

If we look at the data in the tables we can see the relationship is
Customer -> Invoice -> InvoiceLine -> Track -> Album

Since we know the relationship between the tables from the image on brightspace, we can create a sql query easier. We will create the SQL query that will join the five tables and display the following in a table: Customer first and last name, the track name, and the album title.

```python
df = pd.read_sql_query("""
SELECT 
    Customer.LastName,
    Customer.FirstName,
    Track.Name AS TrackName,
    Album.Title AS AlbumTitle
FROM Customer
JOIN Invoice ON Customer.CustomerId = Invoice.CustomerId
JOIN InvoiceLine ON Invoice.InvoiceId = InvoiceLine.InvoiceId
JOIN Track ON InvoiceLine.TrackId = Track.TrackId
JOIN Album ON Track.AlbumId = Album.AlbumId
ORDER BY Customer.LastName, Customer.FirstName;
""", conn)
```

We can now use .head to display the table. We also sorted the names by lastname, firstname
