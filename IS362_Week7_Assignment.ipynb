Starting with importing the pandas library

```python
import pandas as pd
```

Lets load the dataset from my GitHub account to ensure this can be run without saving the file locally and display the data

```python
df = pd.read_csv('https://github.com/KcirtapdiaZ/IS_362_Data_Acquisition_and_Mgmt/raw/main/IS362_Week7_Assignment.csv')
df
```

The first column are the names so we will use this as the index. This will make it easier to the make calculations. Also to make it look cleaner we'll rename the index

```python
df = df.set_index('Unnamed: 0')
df.index.name = 'User'
df
```

Since the indexing is done, we will start with calculating the average rating for each movie. To calculate this we'll use .mean so calculate the average of each movie.

```python
movie_avg = df.mean()
print ("Average rating for each movie:")
print (movie_avg)
```

Now we'll do the same for the average rating for each user

```python
user_avg = df.mean(axis=1)
print ("Average rating for each user:")
print (user_avg)
```

We now need to create a new pandas dataframe with normalized ratings for each user. To do this well remove personal rating bias by subtracting each users average rating from thier ratings. To do this we'll use .sub and axis to make sure the subtraction is occurring by row and target the user average rating

```python
normalized = df.sub(df.mean(axis=1), axis=0)
print ("Normalized ratings:")
print (normalized)
```

Now, we'll show the average ratings for each user and movie with the normalized ratings, again using .mean

```python
print ("Average normalized rating for each movie:")
print (normalized.mean())

print ("Average normalized rating for each user:")
print (normalized.mean(axis=1))
```

Using normalized ratings make it easier to compare users because it adjust for differences in how people rate movies. Some people may rate movies high constantly, while some may rate movies low consistantly. Normalization would help by centering each users ratings. A drawback would be that normalized values do not refelct the original rating scale, in this case 1-5.
